Conversion Rate Analysis
=======
*Note: This IPython Notebook contains code meant to walk through the plots created an article written [here](http://google.com)*

First we have to load some packages for working with the data and graphing it.
I used Python's port of ggplot (written as an R package originally), because I'm familiar with it from using R, and becuase there didn't seem to be an obvious choice of graphing library for Python (Matplotlib, maybe?)

```python
# load python package dependencies
import pandas as pd
import numpy as np
import statsmodels.api as sm
from ggplot import *
# show plots inline instead of in a new window
%matplotlib inline
```

Load the raw data collected from Squawka

```python
# load the data from the csv as a pandas data frame
data = pd.read_csv('conversion_cleaned.csv')

# examine the dataset by viewing the first 5 players
data.head(5)
```

Here the variables of interest are 'crate' (2012 Conversion), and 'crate_next' (2013 conversion. We can generate the first plot from the data right away

```python
# plot conversion rate 2012 - 2013, with red line showing 100% persistence
ggplot(aes(x='crate', y='crate_next'), data=data) + geom_point(size = 40, alpha = 0.8) \
    + xlab("2012 Conversion Rate") + ylab("2013 Conversion Rate") \
    + geom_abline(slope = 1, intercept = 0, colour = "red", linetype = "dotted") 
```

Plotting conversion against shot quality requires breaking out each observation across the two seasons:

```python
# create variables to store conversion rate and shot quality as different observations over the two seasons
allconv = pd.concat([data['crate'],data['crate_next']], ignore_index = True)
allshotq = pd.concat([data['in_area_prop'],data['in_area_prop_next']], ignore_index = True)
# store name + year to label observations
n12 = data['name'] + '12'
n13 = data['name'] + '13'
namelbls = pd.concat([n12,n13], ignore_index = True)
# make a new data frame with these variables
stack = pd.DataFrame({'name': namelbls, 'conv': allconv, 'shotq' : allshotq})
# graph it
ggplot(aes(x='shotq', y='conv'), data=stack) + geom_point() + xlab("Proportion of Shots in Area") + ylab("Conversion Rate")
```

To make the third plot we need to do a little data manipulation, and run regressions for the two separate fit lines

```python
# since the data only has a player's team in each year, we want to define a variable
# which indicates if a player changed teams, called tmchange, which compares whether or
# not a player's team was the same across the two seasons in the sample
data['tmchange'] = [(0 if x ==y else 1) for x,y in zip(data['team'],data['team_next'])]

# we can also make a version of this variable which is labeled more cleanly for plotting
data['Team Change'] = [("Changed Teams" if x ==1 else "Did Not Change Teams") for x in data['tmchange']]

# ggplot adds confidence intervals to linear fit lines by default, and in python there seems to be no
# way to turn them off. so in order to produce the graph with trend lines which splits observations up
# into players who did/did not switch teams, we have to run regressions for each set of points and plot
# the fit lines on the scatterplot
chgonly = data[data['tmchange'] == 1]
stayonly = data[data['tmchange'] == 0]
#onescol = np.ones((len(df),))
coeffs = []
for df in [chgonly,stayonly]:
    Y = df['crate_next'] 
    X = df['crate']
    X = sm.add_constant(X)
    result = sm.OLS(Y, X).fit()
    # uncomment this line to see the regression results
    #print(result.summary())
    # store coefficients
    coeffs.append(result.params)

# plot
# unfortunately, ggplot for python isn't fully supported in IPython 
# notebooks, which I hadn't realized until writing the code here, so the legend gets cut off.
# The blue line/points are players who did not change teams. To get a full version of the plot,
# run the code in a terminal session of IPython or another interpreter.
ggplot(aes(x='crate', y='crate_next', colour = 'Team Change'), data=data) \
    + geom_point(size = 40, alpha = 0.8) + xlab("2012 Conversion Rate") \
    + ylab("2013 Conversion Rate") + geom_abline(slope = coeffs[0][1], intercept = coeffs[0][0], colour= "#f8766d") \
    + geom_abline(slope = coeffs[1][1], intercept = coeffs[1][0], colour= "#00bfc4")
```

Finally, statsmodels gives us the ability to do a straighforward multiple regression on the dataset
containing only players who remained on the same team across both seasons:

```python
Y = chgonly['crate_next'] 
X = chgonly[['crate', 'in_area_prop_next']]
# we have to add a column of ones to the matrix X containing the variables we're using as indpendent variables
# in order to estimate the constant parameter.
X = sm.add_constant(X)
result = sm.OLS(Y, X).fit()
print(result.summary())
```

Plotly graphics
------
To make interactive plotly charts, you'll have to sign up for an account on Plotly's website and get an API key. 
Check out the Python API reference for more information. Here's code for some of the plots I generated. You can view the plots here, but this code won't run unless you sub in your own username and API key:

```python
# plotly graphs
import plotly.plotly as ply 
from plotly.graph_objs import *
# here you'll have to use your own username and API key
ply.sign_in("user", "key")

# simple scatter with player names
trace1 = Scatter(x=data['crate'],y=data['crate_next'], text = data['name'], mode = 'markers', marker = Marker(color = 'blue', size = 10))
gdata = Data([trace1])
layout = Layout(hovermode = 'closest', xaxis = XAxis(title="2012 Conversion Rate"), yaxis = YAxis(title= "2013 Conversion Rate"))
fig = Figure(data = gdata, layout = layout)
ply.plot(fig)

# conversion vs. shot quality scatter:
trace2 = Scatter(x=stack['shotq'],y=stack['conv'], text = stack['name'], mode = 'markers', marker = Marker(color = 'blue', size = 10))
gdata2 = Data([trace2])
layout2 = Layout(hovermode = 'closest', xaxis = XAxis(title="Shot Quality (% of shots in area)"), yaxis = YAxis(title= "Conversion Rate"))

fig2 = Figure(data = gdata2, layout = layout2)
ply.plot(fig2)


# scatter plot of conversion broken out by team change
smtrace = Scatter(x = stayonly['crate'], y = stayonly['crate_next'], text = stayonly['name'], mode = 'markers', marker = Marker(color = 'blue', size = 10), name = "Did not Change Teams")

chgtrace = Scatter(x = chgonly['crate'], y = chgonly['crate_next'],text = chgonly['name'], mode = 'markers', marker = Marker(color = 'orange', size = 10), name ="Changed Teams")
gdata3 = Data([chgtrace,smtrace])
fig3 = Figure(data = gdata3, layout = layout)
ply.plot(fig3)
```
